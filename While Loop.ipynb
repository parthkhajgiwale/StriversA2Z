## Pattern 1 — Counter Controlled While Loop

### Used when moving from A → B but don’t want for.

```python
i = 0
while i<=5:
    print(i)
    i+=1
```

## Pattern 2 — Sentinel / Unknown Input Loop

### Run until user gives a special value.

```python
text = ""
while text != "stop":
    text = input("Enter Something: ")
```

## Pattern 3 — Accumulator While Loop (sum, count, average)

```python
total = 0
i = 1
n = 10
while i <= n:
    total += i
    i += 1
print(f"Average of first {n} numbers: {total//n}")
```

## Pattern 4 — Digit Extraction Loop (VERY IMPORTANT)

### This appears in MANY coding questions.

```python
num = 5432
total = 0
while num > 0:
    digit = num%10
    total +=digit
    num = num//10
print(total)
```

```python
num = 5432
reverse = []
while num > 0:
    digit = num%10
    reverse.append(digit)
    num = num//10
reverse
```

```python
num = 9876
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
rev
```

```python
num = 5342 # You can’t convert an integer directly to a list. Convert to string first, then to list.
print(list(str(num)))

lst = [int(d) for d in str(num)] #If you want integers, not strings
print(lst)        
```

## Pattern 5 — Two Pointer / Condition Movement Loop

```python
arr = [1, 2, 3, 4, 6]
target = 6
left = 0
right = len(arr)-1
while left < right:
    s = arr[left] + arr[right]
    if s==target:
        print(arr[left], arr[right])
        break
    elif s < target:
        left +=1
    else: 
        right-=1
```

```python
vowels = ['a','e','i','o','u']
str0 = "leetcode"
str1 = list(str0)
left = 0
right = len(str1)-1
while left < right:
    if str1[left] in vowels and str1[right] in vowels:
        str1[left], str1[right] = str1[right], str1[left]
        right -=1
        left+=1
    elif str1[left] in vowels and str1[right] not in vowels:
        right-=1
    elif str1[left] not in vowels and str1[right] in vowels:
        left+=1
    else:
        left+=1
        right-=1
        
print(''.join(str1))
```

## Pattern 6 — Infinite Loop with Break (VERY INTERVIEW USEFUL)

```python
while True:
    x = input("Enter exit to stop")
    if x == "exit":
        print("Infinite Loop exited")
        break
```
